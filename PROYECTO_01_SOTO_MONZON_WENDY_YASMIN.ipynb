3. Calcula la media, máximo, mínimo y desviación estándar de las columnas con valores numéricos.

```python
#Para esto se puede hacer de dos maneras la mas laboriosa seria calculandolo de uno por uno, así como se muestra a continuación

#Se calculan la media, el máximo, el mínimo y la desviación estándar de la columna Age
print('La media de la columna Age es: ', df['Age'].mean())
print('El maximo de la columna Age es: ', df['Age'].max())
print('El minimo de la columna Age es: ', df['Age'].min())
print('La desviacion estandar de la columna Age es: ', df['Age'].std())

#Se calculan la media, el máximo, el mínimo y la desviación estándar de la columna Income
print('La media de la columna Income es: ', df['Income'].mean())
print('El maximo de la columna Income es: ', df['Income'].max())
print('El minimo de la columna Income es: ', df['Income'].min())
print('La desviacion estandar de la columna Income es: ', df['Income'].std())

```
